#### Question 40
Which preferred position type of England has on average the highest overall rating? (2 marks)

```python
football[football['Nationality'] == 'England'].groupby('Preferred Positions Type').agg(avg = ('Overall', 'mean')).sort_values(by = 'avg', ascending= False)
```

#### Question 41
Brazil’s forwards have a higher average overall rating than the backs (1 mark)

```python
football[football['Nationality'] == 'Brazil'].groupby('Preferred Positions Type').agg(avg = ('Overall', 'mean')).sort_values(by = 'avg', ascending= False)
```

#### Question 43
Which of the following attributes is on average the lowest for goalkeepers? (3 marks)

#### Question 44
Which preferred positions type has the most entries in this dataset? (3 marks)

```python
football.groupby('Preferred Positions Type').agg(n = ('Preferred Positions Type','count')).sort_values(by = 'n', ascending= False)
```

#### Question 45 
Which player from Portugal, that is younger than 25, has the highest overall rating? (3 marks)

```python
football[(football['Nationality'] == 'Portugal') & (football['Age'] < 25)][['Name', 'Age', 'Nationality', 'Overall']].sort_values(by= 'Overall', ascending= False)
```
